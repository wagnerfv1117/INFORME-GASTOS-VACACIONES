- Grafico para comparar las variables de datos en los año 2022 y 2023

```python
x1 = df1['MONTO']

x2 = df2['MONTO']


fig, ax = plt.subplots()

ax.plot(x1,  label='Gastos año 2022', color= 'blue')
ax.plot(x2,  label='Gastos año 2023', color='orange')

ax.set_title('COMPRATIVO DE GASTOS EN VACACIONES 2022 vs 2023')
ax.set_xlabel('Posición de los datos en el DataFrame')
ax.set_ylabel('Unidades monetarias - (COP)')

ax.legend()

plt.show()

plt.savefig("C:/Users/wagne/Downloads/ANALISIS GASTOS VACACIONES/reports/comparativo.jpg")# ejemplo para guardar imagen en la carpeta reportes.
```
